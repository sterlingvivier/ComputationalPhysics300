# Polynomial Regression from Scratch

For a second order polynomial, we have the below equation 

\begin{equation*}
 f(x) = ax^2 + bx + c
\end{equation*}

where a, b, and c are unknowns

In order to find these unknowns, we can set up the following matrices, A, b, and x


The below code takes two matrixes A and b to find the coefficients of our polynomial function. Matrix A takes the below form 
\begin{equation*}
A = 
\begin{pmatrix}
\Sigma x^0 & \Sigma x^1 & \Sigma x^2 \\
\Sigma x^1 & \Sigma x^2 & \Sigma x^3 \\
\Sigma x^2 & \Sigma x^3 & \Sigma x^4
\end{pmatrix}
\end{equation*}

Matrix x takes the below form

\begin{equation*}
x = 
\begin{pmatrix}
a \\
b\\
c
\end{pmatrix}
\end{equation*}

Matrix b is then defined as 

\begin{equation*}
b = 
\begin{pmatrix}
\Sigma x^0*y_i \\
\Sigma x^1*y_i \\
\Sigma x^2*y_i 
\end{pmatrix}
\end{equation*}

We can find our coefficients a, b , and c by the formula 

\begin{equation*}
x = A^{-1} * b
\end{equation*}

1. first generating a random set of data && plotting it below

```python
import numpy as np
import matplotlib.pyplot as plt

#generating random data && plotting
np.random.seed(2)
x = np.random.rand(500,1)
y = 5*((x)**(2)) + np.random.rand(500,1)
fig = plt.figure(figsize=(8,6))
plt.scatter(x, y, color= 'y')
plt.legend(["Data"])
plt.title("Random Data")
```

when n = 1

```python
n=1 #order of degree
A = np.zeros((n+1, n+1)) #defines A matrix
b = np.zeros((n+1, 1)) #defines b matrix
np.random.seed(2)
x = np.random.rand(500,1)
y = 5*((x)**(2)) + np.random.rand(500,1)

for i in range(n+1): #places values in our two arrays
    for j in range(n+1):
        A[i, j] = sum(x**(i+j))
    b[i] = sum(x**(i)*y)


c = np.dot(np.linalg.inv(A), b) #takes the inverse && multiplies
c = np.flip(c)
c = c.ravel()

f=np.poly1d(c) #fits a function with our coefficients
print(f) #outputs the function

x2 = np.arange(0, 1, .01) #x-array to plot function

def funplot(x):
    f= 5.013*x - 0.3746
    return f

fig = plt.figure(figsize=(8,6)) #plotting our data along with our regression
plt.plot(x2, funplot(x2), 'black')
plt.scatter(x, y)
plt.legend(["n=1 fit", "data"])
plt.title("n=1")
```
