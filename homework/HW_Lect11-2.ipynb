```python
import matplotlib.pyplot as plt
import numpy as np
from random import random


    # program to calculate the area in between 
    # the two curves using monte carlo method


fig1 = plt.figure()
ax1 = fig1.add_subplot(111, aspect='equal')

x1 = np.linspace(0,2,100)
y1 = np.sqrt(1-(x1-1)**2) #first curve
ax1.plot(x1,y1,'k')

y2 = 2-np.sqrt(4-x1**2) #second curve
ax1.plot(x1,y2,'k')

ax1.set_xlim(0,2)
ax1.set_ylim(0,2)
ax1.fill_between(x1, y1, y2, where=y1>y2, facecolor='green')

plt.show()

#functions for monte carlo simulation
def f1(x):
    return np.sqrt(1-(x-1)**2)
def f2(x):
    return 2-np.sqrt(4-x**2)

#monte carlo function
def MC(N):
    count = 0
    for i in range(N):
        x = 1.6*random()
        y = random()
        if ((y<f1(x)) & (y>f2(x))):
            count += 1
    I = 1.6*count/N
    return I

#different monte carlos printed for different 
#amounts of iterations
narr = np.arange(100, 10000, 100)
mcarr = []

for n in range(100, 10000, 100):
    mcarr.append((MC(n)))
    
plt.plot(narr, mcarr) #plot how MC changes with # of iterations

print('different MC results: ')
print('N=100:    ', MC(100))
print('N=1000:   ',MC(1000))
print('N=10000:  ',MC(10000))
print('N=100000: ',MC(100000))

```
