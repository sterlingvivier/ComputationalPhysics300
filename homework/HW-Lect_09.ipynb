The data I chose to use for the Fourier analysis came from the textbook "Computational Physics Problem Solving with Computers." The below code first defines a "signal" function, plots it, and then uses this signal function
as a parameter for our Fourier Analysis. Once the analysis is peformed, the new results are also plotted. 

```python
import numpy as np
import matplotlib.pyplot as plt
import cmath                            # complex math
    
    
N = 500;                  Np = N                           
signal = np.zeros( (N+1), float )     
xa = np.zeros( (N+1), float )   

def f(signal):                          # Signal
    step = 2*np.pi/N;        
    x = 0. 
    for i in range(0, N+1):
        signal[i] = 30*np.cos(x*x*x)
        xa[i] = x                      # Plot
        x += step
        
f(signal)
plt.plot(xa, signal)
plt.show()
plt.close()

yf = np.fft.fft(signal)
plt.plot(xa, yf)
plt.show()
```
