```python
# HW-Lec_03 Sterling Vivier PHYS 300
from math import log, exp             #importing libraries
import matplotlib.pyplot as plt 


#defining a function that calculates the factorial
#parameter: number we want the factorial of
#error checks parameter & returns the total
def factor(N):                       
    N = int(N)
    total = 1
    if N<0:
        print('Error, N must be greater than 0')
    elif N>1: 
        for i in range(2,N+1):
            total *= i
    return total

#defines a function that calculates the multiplicity/omega
#parameters are number of oscillators & energy
def Multiplicity(N,n):
    #return factor(N)/factor(n)/factor(N-n)
    return exp(log(factor(N)) - log(factor(n)) - log(factor(N-n)))

#constants for this problem set
Na = 300
Nb = 200
q_total = 100
qa_list = range(q_total)
omega_total = []

for qa in qa_list:
    #subtracts qa from the given total
    qb = q_total - qa
    #calculates the two omegas
    omega_a = Multiplicity(Na, qa)
    omega_b = Multiplicity(Nb, qb)
    #appends to omega array
    omega_total.append(omega_a*omega_b)

#plots output
plt.plot(qa_list, omega_total)    
plt.xlabel('$q_a$')
plt.ylabel('$\Omega$')
plt.show()
```

```python
#utilizes the same previously defined functions
#just now with different inputs
Na = 3000
Nb = 2000
q_total = 100
qa_list = range(q_total)
omega_total = []

for qa in qa_list:
    qb = q_total - qa
    omega_a = Multiplicity(Na, qa)
    omega_b = Multiplicity(Nb, qb)
    omega_total.append(omega_a*omega_b)
plt.plot(qa_list, omega_total)    
plt.xlabel('$q_a$')
plt.ylabel('$\Omega$')
plt.show()

#energy q stayed the same in both problem sets, but number of oscillators increased. 
#the distribution still peaks in the same range qa = 40-80, however omega is increased.
#the first plot shows the peak at 1.75e106, while the second plot shows it at 2.5e210.
#the order of magnitude almost doubled when the order of magnitude for the inputs was
#increased by 1. 
```
